```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.linear_model import LinearRegression
import statistics
import math
import scipy.stats as stats
import statsmodels.stats.stattools as stools
import statsmodels.stats as stats 
import matplotlib.pyplot as plt

import statsmodels.api as sm
import pylab 
```

```python
# Data from all the different economic data, atleast until now.

Gold_prices_Adjusted_inflation = pd.read_csv("Gold-annual-prices-inflation-Adjusted.csv")
ten_year_treasury_rate_annual_average_yield = pd.read_csv("10-year-treasury-rate-annual-average-yield.csv")
VIX_data = pd.read_csv("VIX-annual-average_1991-2020.csv")
world_gdp_growth_rate = pd.read_csv("world-gdp-growth-rate.csv")
world_gdp_gross_domestic_product = pd.read_csv("world-gdp-gross-domestic-product.csv")
world_inflation_rate = pd.read_csv("world-inflation-rate-cpi.csv")
```

```python
# Next it is necessary to place all the necessary data from all variables in the same
# dataframe, from 1991 to 2020.

years = Gold_prices_Adjusted_inflation.iloc[1:31,0]
years = years.iloc[::-1]
years.reset_index(drop=True,inplace=True)

# gold prices adjusted for inflation.
GPAI = Gold_prices_Adjusted_inflation.iloc[1:31,1].reset_index(drop=True)
GPAI = GPAI.iloc[::-1]
GPAI.reset_index(drop=True,inplace=True)

# 10 year treasury rate annual average yield
TYTRAAY = ten_year_treasury_rate_annual_average_yield.iloc[1:31,1]
TYTRAAY = TYTRAAY.iloc[::-1]
TYTRAAY.reset_index(drop=True,inplace=True)

# VIX, Cboe Volatity Index.
VIX = VIX_data.iloc[1:31,1]
VIX = VIX.iloc[::-1]
VIX.reset_index(drop=True,inplace=True)

# GPD rate(%).
GDP = world_gdp_growth_rate.iloc[30:60,1]
GDP.reset_index(drop=True, inplace=True)

# World inflation rate(%).
WIR = world_inflation_rate.iloc[10:40,1]
WIR.reset_index(drop=True, inplace=True)

# the new aggregated dataset, the main dataset from the analysis.
df = pd.concat([years,GPAI,TYTRAAY,VIX,GDP,WIR],axis=1)
df
```

```python
sns.lineplot(x = df.iloc[:,0], y = df.iloc[:,1])
```

```python
# trying to fit a good multiple Linear Regression model to represent the data.
# independent variables without years and VIX
x = df.iloc[:,[2,4,5]]

# dependent variables.
y = df["Gold Year Close"]

# regression modeland fit.
reg = LinearRegression()
reg.fit(x,y)

# Return the coefficient of determination R^2 of the prediction.
print("\ncoefficient of determination R^2:",reg.score(x,y),"\n")

# model coefficients:
print("model coefficients:",reg.coef_,"\n")

#model intercept:
print("model intercept:",reg.intercept_,"\n")
```

```python
x = sm.add_constant(x)
mod = sm.OLS(y,x)
res = mod.fit()
print(res.summary())
#print(res.predict())
#print(res.params)
```

```python
# predicted values:
predicted = res.predict()

# residuals:
residuals = GPAI - predicted

# multiple linear regression graph: (adjusted and predicted)
predicted = res.predict()

sns.scatterplot(x = df.iloc[:,0],y=y)
sns.scatterplot(x=df.iloc[:,0],y=predicted,color="red")
```

```python
# residual analysis:

# leverage calculation. First manual using the formula: H = X'*(X*X')^-1*X
a = np.array(df.iloc[:,[2,4,5]])
H = np.dot(np.dot(a,np.linalg.inv(np.dot(a.transpose(),a))),a.transpose())
leaverage = H.diagonal()

influence = res.get_influence()

# leverage calculation automatic:
leverage_python = influence.hat_matrix_diag

# standardized and studentized residuals.
standardized_residuals = influence.resid_studentized_internal
studentized_residuals = influence.resid_studentized_external

# cook's distance for each observation.
cooks = influence.cooks_distance
```

```python
# scatterplot (residuals vs dependent variable).
sns.scatterplot(x= y, y= residuals)
```

```python
# residuals plot (residuals vs predicted values).
sns.scatterplot(x = predicted,y = residuals)
```

```python
# standardized residuals plot.
sns.scatterplot(x = predicted,y = standardized_residuals)
```

```python
# studentized residuals plot.
sns.scatterplot(x = predicted,y = standardized_residuals)
```

```python
# influence plot.
sm.graphics.influence_plot(res,criterion="cooks",alpha=0.5)
```

```python
# leverage plot:
u = statistics.mean(leverage_python)
i = np.arange(0,30,1)

graph = sns.scatterplot(x = i,y = leverage_python)
graph.axhline(2*u,color="green")
```

```python
# cooks distance plot:
u = statistics.mean(cooks[0])
i = np.arange(0,30,1)

graph = sns.scatterplot(x = i,y = cooks[0])
```
